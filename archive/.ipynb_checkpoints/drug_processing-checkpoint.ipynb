```python
import pandas as pd
path = 'data/GDSC1_fitted_dose_response_25Feb20.csv'
df = pd.read_csv(path)
df.head()
```

```python
df_cleaned = df.filter(['DRUG_NAME', 'PUTATIVE_TARGET', 'PATHWAY_NAME', 'AUC'])
df_cleaned.head()
```

```python
# OPTION 1: drop all of the duplicates and keep the mean AUC for each unique entry under 'DRUG_NAME'
df_final = pd.DataFrame(df_cleaned.groupby(['DRUG_NAME', 'PUTATIVE_TARGET', 'PATHWAY_NAME'], as_index=False)['AUC'].mean())
df_final.head()
```

```python
# OPTION 2: drop all of the duplicates and keep the first AUC
cols = ['DRUG_NAME', 'PUTATIVE_TARGET', 'PATHWAY_NAME']
df_final = df_cleaned.drop_duplicates(cols)
df_final
```
