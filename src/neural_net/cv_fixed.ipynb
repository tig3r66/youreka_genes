# Helper Functions

## K-Folds Evaluation Function

```python
def cv_eval(build_fn, params, X, y, n_folds=5, base_fpr=np.linspace(0, 1, 101)):
    """K-fold validation function for evaluation of a Keras model.
    Arguments:
        build_fn: a valid build function.
        params (dict): a dictionary of parameters for the model.
        n_folds (int): the number of cross-folds.
        X (pandas dataframe): values of the feature variables.
        y (pandas dataframe): values of the classification.
    Returns:
        cms (list): confusion matrix for each cross-fold.
        histories (list): model metrics for each cross-fold.
        tprs (list): the True Positive Rate for each cross-fold.
        threshes (list): the thresholds for each cross-fold.
    """
    kf = StratifiedKFold(n_splits=n_folds, shuffle=False)
    cms = []  # confusion matrix results
    histories = []  # for model analysis
    tprs = []  # for ROC analysis
    threshes = []

    for train, test in kf.split(X, y):
        # creating the model
        model = KerasClassifier(build_fn=build_fn,
                                epochs=params['epochs'],
                                batch_size=params['batch_size'],
                                neurons=params['neurons'],
                                optimizer=params['optimizer'],
                                kernel_regularizer=params['kernel_regularizer']
                               )
        X_train, X_test = X.iloc[train].values, X.iloc[test].values
        y_train, y_test = y.iloc[train].values, y.iloc[test].values
        # recording model metrics and confusion matrix
        histories.append(model.fit(X_train, y_train, validation_data=(X_test, y_test), verbose=0))
        cms.append(confusion_matrix(y_test, model.predict(X_test)))
        # recording for auc
        y_score = model.predict_proba(X_test)
        fpr, tpr, thresh = roc_curve(y_test, y_score[:, 1])
        threshes.append(thresh)  # getting thresholds
        tpr = np.interp(base_fpr, fpr, tpr)  # interpolating
        tprs.append(tpr)  # getting true positive rate

    return cms, histories, tprs, threshes
```

## Confusion Matrix Plotting Function

```python
def plot_cm(cm, target_names, title='Confusion matrix', cmap=None, normalize=True, cbar=True):
    """Plots a confusion matrix based on a nested array.
    Arguments:
        cm (list): square matrix with classification counts.
        target_names (list): classifier names.
        title (str): title of the confusion matrix plot.
        cmap (str): colour theme for the plot.
        normalize (bool): plots proportions instead of counts.
        cbar (bool): shows colourbar if True.
    """
    accuracy = np.trace(cm) / np.sum(cm).astype('float')
    misclass = 1 - accuracy

    if cmap is None:
        cmap = plt.get_cmap('Blues')

    fig = plt.figure(figsize=(4, 3), dpi=300)
    ax1 = fig.add_subplot()
#     ax1.text(-.1, 1.1, 'B', transform=ax1.transAxes, fontsize=16, fontweight='bold', va='top', ha='right')
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)

    if target_names is not None:
        tick_marks = np.arange(len(target_names))
        plt.xticks(tick_marks, target_names)
        plt.yticks(tick_marks, target_names, rotation=90, verticalalignment='center')
#         plt.yticks([], ['', ''])

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    thresh = cm.max() / 1.5 if normalize else cm.max() / 2
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        if normalize:
            plt.text(j, i, "{:0.4f}".format(cm[i, j]),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")
        else:
            plt.text(j, i, "{:,}".format(cm[i, j]),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")
    if cbar:
        plt.colorbar()
#         plt.clim(0, 60)

    plt.tight_layout()
    plt.ylabel('True')
    plt.xlabel('Predicted\naccuracy={:0.3f}; misclass={:0.3f}'.format(accuracy, misclass))
    plt.show()

    return fig
```

## Gets Mean ROC Stats - Training

```python
def get_roc_stats(tprs, base_fpr=np.linspace(0, 1, 101), n_folds=5):
    tprs = np.array(tprs)
    mean_tpr = tprs.mean(axis=0)
    std = tprs.std(axis=0)
    tprs_upper = mean_tpr + std / np.sqrt(n_folds)
    tprs_lower = mean_tpr - std / np.sqrt(n_folds)
    temp_auc = auc(base_fpr, mean_tpr)
    return mean_tpr, std, tprs_upper, tprs_lower, temp_auc
```

## Creates Data for ROC Curves - Testing

```python
def make_roc(model, X, y):
    base_fpr = np.linspace(0, 1, 101)
    y_score = model.predict_proba(X)
    fpr, tpr, thresholds = roc_curve(y_test, y_score[:, 1])
    tpr = np.interp(base_fpr, fpr, tpr)
    return tpr, thresholds
```

## Getting Threshold Based on Youden's Index (for use with testing data)

```python
def get_youden_j(fpr, tprs, thresholds):
    """Gets Youden's index from ROC curve.
    """
    threshes = []
    j_inds = []
    for i in range(len(thresholds)):
        j_scores = tprs[i] - fpr
        j_sorted = sorted(zip(j_scores, thresholds[i]))
        j_inds.append(j_sorted[-1][0])  # Youden's index
        threshes.append(j_sorted[-1][1])  # threshold
    return np.mean(j_inds), np.mean(threshes)
```

## Using Youden's Index (From Training Data) as Threshold

```python
def get_classes(model, X, j_idx):
    """Predicts binary classes based on a threshold.
    Arguments:
        model (KerasClassifier): the trained model.
        X (pandas dataframe): the testing data.
        j_idx (float): threshold for classification.
    Returns:
        classes (list): a list of predicted classes.
    """
    probas = model.predict_proba(X.values)
    classes = []
    for i in probas:
        if i[0] >= j_idx:
            classes.append(0)
        else:
            classes.append(1)
    return classes
```
