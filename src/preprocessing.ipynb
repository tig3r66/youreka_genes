```python
import pandas as pd
```

```python
path = 'data/GDSC2_fitted_dose_response_25Feb20.csv'
```

```python
df = pd.read_csv(path)
df.head()
```

```python
cleaned_table = pd.pivot_table(df, columns="DRUG_NAME", index="CELL_LINE_NAME", values="LN_IC50")
```

```python
clean_df = pd.DataFrame(cleaned_table)
clean_df.head()
```

```python
clean_df.to_csv('cleaned.csv')
```
